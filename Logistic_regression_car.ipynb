```
import io
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats
import scipy.special
import seaborn as sns
```

```
def one_hot(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode 
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```
df = pd.read_csv("Navin_Vehicle_data.csv")
df.head()
```

```
df.describe()
```

```
df = one_hot(df, ["model"])
df.head()
```

```
df['range']=(df['range']-df['range'].mean())/df['range'].std()
df.describe()
```

```
X=df[['range','model_amaze-diesel','model_amaze-petrol',	'model_city-diesel',	'model_city-petrol',	'model_jazz-diesel',	'model_jazz-petrol']].values
Y=df['wheel_alignment_and_balancing'].values
```

```
# Import 'LogisticRegression' and create a LogisticRegression object
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X,Y)
```

```
from sklearn.metrics import confusion_matrix
y_pred = logreg.predict(X[:200])
confusion_matrix = confusion_matrix(Y[:200], y_pred)
print(confusion_matrix)
```

```
from sklearn.metrics import classification_report
print(classification_report(Y[:200], y_pred))
```
